```python
import pandas as pd
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import StandardScaler, LabelEncoder, MinMaxScaler
from sklearn.metrics import accuracy_score, mean_absolute_error
```

```python
ruta_datos = "datos/data.csv"
df = pd.read_csv(ruta_datos)
print(df.shape)
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["Unnamed: 32"], inplace=True)
df.info()
```

```python
df.duplicated().sum()
```

```python
df['diagnosis'].value_counts()
```

```python
df.head(3)
```

```python
colum_object = df.select_dtypes(include=['object']).columns
colum_object
```

```python
for col in colum_object:
    le = LabelEncoder()
    df[col+"le"] = le.fit_transform(df[col])
    print(f"columna: {col} - clases: {le.classes_}")

df.head(3)
```

## 2. Selección de matriz "X" o estimulos y objetivo o "y"

```python
lista_columnas = df.select_dtypes(include=['float64', 'int64']).columns
lista_columnas
```

```python
df_final = df[lista_columnas]
df_final.info()
```

```python
x = df_final.drop(columns=["diagnosisle"])
y = df_final["diagnosisle"]

print(x.shape)
print(y.shape)
```

```python
x.to_csv("dato.csv", index=False)
```

## 3. Entrenar un red neuronal
1. Cortar datos en prueba y entrenamiento
2. Proponer un modelo o arquitectura de redes neuronales
3. Entrenar el modelo

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape[1]
```

```python
model = Sequential([Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
                    Dense(32, activation='relu'),
                    Dense(1, activation='sigmoid')])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
ruta_guardar = "modelos/modelo_cancer_mama.keras"
model.save(ruta_guardar)
```
