```python
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([Dense(64, activation='relu', input_shape=(31,)),
                    Dense(32, activation='relu'),
                    Dense(1, activation='sigmoid')])
```

```python
model.load_weights("modelos/modelo_cancer_mama.keras")
```

```python
dato=pd.read_csv("dato.csv")
```

```python
predicciones = model.predict(dato.iloc[0:1])
print(predicciones)
```
